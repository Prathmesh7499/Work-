## Objectives



 - Use Webscraping to extract required information from a website.
 - Use Pandas to load and process the tabular data as a dataframe.
 - Use Numpy to manipulate the information contatined in the dataframe.
 - Load the updated dataframe to CSV file.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


url = 'https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29'

tables = pd.read_html(url)
df = tables[3]
df
```

```python
df.columns = range(df.shape[1])
df
```

### Select two column for analysis

```python
df = df[[0,2]]
df
```

```python
df = df.iloc[1:11,:]
df
```

### Assign name of the column 

```python
df.columns = ['Country','GDP(USD)']
df
```

### Change the data type of the 'GDP (Million USD)' column to integer. Use astype() method.


```python
df['GDP(USD)'] = df['GDP(USD)'].astype(int)
df
```

### Convert the GDP value in Million USD to Billion USD

```python
df[['GDP(USD)']] = df[['GDP(USD)']]/1000
df
```

```python
df[['GDP(USD)']] = np.round(df[['GDP(USD)']],2)
df
```

### Rename the column name 

```python
df.rename(columns = {'GDP(USD)' : 'GDP(Billion_USD)'})
```

### load the dataframe to CSV file

```python
df.to_csv('./Largest_economies.csv')
df
```
